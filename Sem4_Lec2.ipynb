# Second Order ODE

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def f(u, t):
    y, v = u       # Unpacking
    f1 = v
    f2 = -lam*v - k*y
    return [f1, f2]

lam , k = 0.5, 2
u0 = [0, 1]
t = np.linspace(0, 20, 1001)
odeint(f, u0, t)
```

```python
sol = odeint(f, u0, t)
y = sol[:, 0]
v = sol[:, 1]
plt.plot(t, y, lw = 3, c = 'r')
plt.xlim(0, 20)
plt.axhline()
plt.show()
```

```python
plt.plot(t, v, lw = 3, c = 'r')
plt.xlim(0, 20)
plt.axhline()
plt.show()
```
