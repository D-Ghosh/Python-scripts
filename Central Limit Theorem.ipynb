```python
def udev(x):
    a, c, m = 2416, 374441, 1771875
    x = (a*x + c) % m
    return x, x/m
```

```python
def udevs(x, size):
    X = []
    for i in range(size):
        x, r = udev(x)
        X.append(r)
    return(X)
```

```python
udevs(123, 10000)
```

```python
L = udevs(123, 100000)
import matplotlib.pyplot as plt
plt.hist(L, bins = 40, edgecolor = 'black')
```

```python
import numpy as np
import matplotlib.pyplot as plt

sample = lambda: np.mean(np.random.uniform(size = 100))
x = [sample() for _ in range(10000)]

plt.hist(x, bins = 40, edgecolor = 'black')
```

```python
x = lambda n: np.mean(np.random.randint(1, 7, size = n))
k = 1
for n in [1, 2, 5, 10]:
    xx = [x(n) for _ in range(10000)]
    plt.subplot(2, 2, k)
    plt.hist(xx, bins = 40, density = True, edgecolor = 'black')
    k += 1
plt.show()
```
