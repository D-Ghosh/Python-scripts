```python
from scipy import special as s
n, x =5, 0.75
Pn = s.legendre(n)
Pn_1 = s.legendre(n-1)
Pn_2 = s.legendre(n-2)
LHS = n*Pn(x)
RHS = (2*n-1)*x*Pn_1(x) - (n-1)*Pn_2(x)
print(LHS == RHS)
```
