```python
from scipy.signal import square
import numpy as np
t = np.linspace(0, 4*np.pi, 1001)
import matplotlib.pyplot as plt
plt.plot(t, square(t), '.')
plt.show()
```

```python
t = np.linspace(0, 1, 1001)
plt.plot(t, square(t), '.')
plt.show()
```

```python
plt.plot(t, square(2*4*np.pi*t))
plt.show()
```

```python
plt.plot(t, square(2*4*np.pi*t, duty = 0.6))
plt.show()
```

```python
from scipy.signal import sawtooth
help(sawtooth)
```

```python
t = np.linspace(0, 4*np.pi, 1001)
plt.plot(t, sawtooth(t))
plt.show()
```

```python
t = np.linspace(0, 4*np.pi, 1001)
plt.plot(t, sawtooth(t, width = 0.5))
plt.show()
```

```python
t = np.linspace(0, 8*np.pi, 1001)
plt.plot(t, sawtooth(t, width = 0.8))
plt.show()
```
