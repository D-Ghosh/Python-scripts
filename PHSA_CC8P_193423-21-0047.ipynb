```python
# Q4(i)

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 101)
u = np.zeros(101)
u[0] = 1

for j in range(100):
    for i in range(1, 100):
        u[i] += (u[i+1] - 2*u[i] + u[i-1])/4
        
plt.plot(x, u, lw = 3)
plt.xlabel("Along the length of the rod", size = 18)
plt.ylabel("Temperature", size = 18)
plt.title("Temperature profile when rod is heated at left end", size = 20)
plt.show()
```
